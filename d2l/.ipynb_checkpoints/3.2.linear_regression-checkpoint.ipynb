线性回归的从零开始实行

```python
%matplotlib inline
import random
import torch
import matplotlib.pyplot as plt
from d2l import torch as d2l
```

```python
def synthetic_data(w, b, num_examples):
    """
    w: Tensor (shape: (d,) 或 (d, 1))
    b: Scalar(float), 真实的偏置项
    num_examples, int 要生成的样本数量
    """
    X = torch.normal(0, 1, (num_examples, len(w)))
    y = torch.matmul(X, w) + b
    y += torch.normal(0, 0.01, y.shape)
    return X, y.reshape((-1, 1))
```

`torch.normal(mean, std, size, *, out=None, generator=None)`

- mean （float）分布的均值
- std (float) 分布的标准差
- size (tuple): 输出张量的形状



`torch.matmul` 是 PyTorch 中用于通用矩阵乘法（General Matrix Multiplication）的核心函数，支持多种维度的张量输入，并自动处理广播（broadcasting）。
`torch.matmul(input, other, out=None)`
| 输入维度      | 行为                                                       | 等价函数              |
| --------- | -------------------------------------------------------- | ----------------- |
| 1D × 1D   | 向量点积（内积）                                                 | `torch.dot`       |
| 2D × 2D   | 标准矩阵乘法                                                   | `torch.mm`        |
| 1D × 2D   | 向量与矩阵相乘（向量视为行向量）                                         | —                 |
| 2D × 1D   | 矩阵与向量相乘（向量视为列向量）                                         | `torch.mv`        |
| ≥3D × ≥3D | **批量矩阵乘法**（Batch Matrix Multiplication），最后两维做矩阵乘法，前面维度广播 | `torch.bmm`（仅限3D） |


```python
true_w = torch.tensor([2, -3.4])
true_b = 4.2
print(true_w)
features, labels = synthetic_data(true_w, true_b, 1000)
print('features:', features[0],'\nlabel:', labels[0])
```

```python
d2l.set_figsize()
d2l.plt.scatter(features[:, (1)].detach().numpy(), labels.detach().numpy(), 1)
```

```python
def data_iter(batch_size, features, labels):
    num_examples = len(features)
    indices = list(range(num_examples))
    random.shuffle(indices)
    for i in range(0, num_examples, batch_size):
        batch_indices = torch.tensor(
            indices[i: min(i + batch_size, num_examples)])
        yield features[batch_indices], labels[batch_indices]
```

```python
batch_size = 10
for X, y in data_iter(batch_size, features, labels):
    print(X, '\n', y)
    break
```

```python
"""
初始化模型参数
从均值为0，标准差为0.01的正态分布中采样随机数来初始化权重
将偏置初始化为0
"""
w = torch.normal(0, 0.01, size = (2,1), requires_grad = True)
b = torch.zeros(1, requires_grad = True)
print(w)
print(b)
```

```python
"""
定义模型
"""
def linreg(X, w, b):
    return torch.matmul(X, w) + b
```

```python
"""定义损失函数"""
def squared_loss(y_hat, y):
    return (y_hat - y.reshape(y_hat.shape))**2 / 2
```

| 部分                       | 含义                                            |
| ------------------------ | --------------------------------------------- |
| `y_hat`                  | 模型的**预测值**（$\hat{y}$），即 `linreg(X, w, b)` 的输出 |
| `y`                      | 真实的**标签值**（ground truth）                      |
| `y.reshape(y_hat.shape)` | 将真实值 `y` 重塑为与预测值 `y_hat` **完全相同的形状**，保证维度匹配   |
| `y_hat - y.reshape(...)` | 计算**残差**（预测值与真实值的差距）                          |
| `** 2`                   | 对残差进行**平方**（消除正负号，放大大的误差）                     |
| `/ 2`                    | 除以 2（为了求导后系数简洁，见下文）                           |


本代码计算的是单个样本的损失。如果要得到整个数据集的损失，通常会在外面再套一层`mean()`



```python
"""定义优化算法"""
def sgd(params, lr, batch_size):
    """小批量随机梯度下降"""
    with torch.no_grad():
        for param in params:
            param -= lr * param.grad / batch_size
            param.grad.zero_()
```

```python
"""训练"""
lr = 0.03
num_epochs = 3
net = linreg
loss = squared_loss
for epoch in range(num_epochs):
    for X, y in data_iter(batch_size, features, labels):
        l = loss(net(X, w, b), y)
        l.sum().backward()
        sgd([w, b], lr, batch_size)
    with torch.no_grad():
        train_l = loss(net(features, w, b), labels)
        print(f'epoch {epoch + 1}, loss {float(train_l.mean()):f}')
```

```python
print(f'w的估计误差: {true_w - w.reshape(true_w.shape)}')
print(f'b的估计误差: {true_b - b}')
```
