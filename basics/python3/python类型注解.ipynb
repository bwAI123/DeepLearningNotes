## python 类型注解

`Literal` 是 Python 类型注解系统中的一个"精确值类型"，它表示一个变量不是任意字符串/整数，而只能是某个特定的字面量值。

```python
from typing import Literal

def set_mode(mode: Literal["train", "test"]):
    print(f'mode is {mode}')

set_mode("train")   # 静态检查：✓ 合法
set_mode("hello")   # 静态检查：✗ 报错
                    # 运行时：两种情况都正常执行，不会报错
```

```python
class ModelConfig:
    optimizer: Literal["adam", "sgd", "rmsprop"]
    device: Literal["cpu", "cuda", "mps"]

config = ModelConfig()
config.optimizer = "adam"     # ✓
config.optimizer = "lamb"     # ✗ 类型检查器报错
```

```python
from typing import Literal

def process_data(mode: Literal["train", "test", "val"]):
    print(f"Processing in {mode} mode")

process_data("train")   # ✓ 合法
process_data("test")    # ✓ 合法
process_data("hello")   # ✗ 静态类型检查器（如 mypy）会报错：
                        #   error: Argument 1 to "process_data" has incompatible type "Literal['hello']"
```
