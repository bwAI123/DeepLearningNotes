```python
import numpy as np

# 设置随机种子保证可重复
np.random.seed(2024)

# ========== 参数设定 ==========
x_0 = np.array([100.0, 80.0, 60.0, 40.0])  # 2x2 图像拉平的4维向量
beta_t = 0.1

# 均值向量 μ = sqrt(1-β) * x_0
mu = np.sqrt(1 - beta_t) * x_0

# 协方差矩阵 Σ = β_t * I (4x4)
Sigma = beta_t * np.eye(4)

# ========== 方法1：直接从多元高斯采样 ==========
n_samples = 10
samples = np.random.multivariate_normal(mu, Sigma, size=n_samples)
print(samples)

# ========== 方法2：手动实现（重参数化技巧）==========
# 公式: x_t = sqrt(1-β) * x_{t-1} + sqrt(β) * ε
# 其中 ε ~ N(0, I)，即每个分量独立从标准正态采样
epsilon = np.random.randn(4)          # 4个独立标准正态随机数
x_t_manual = np.sqrt(1 - beta_t) * x_0 + np.sqrt(beta_t) * epsilon
```
