```python
import os
import torch
import torch.nn as nn
from torch.utils.data import TensorDataset, DataLoader
from typing import Tuple, Optional


def synthetic_data(
    w: torch.Tensor,
    b: Union[float, torch.Tensor],
    num_examples: int,
    noise_std: float = 0.01,
    seed: Optional[int] = None,
    device: Union[str, torch.device, None] = None
) -> Tuple[torch.Tensor, torch.Tensor]:
    """
    生成线性回归的合成数据集。
    
    Model: y = Xw + b + epsilon, where epsilon ~ N(0, noise_std^2)
    """
    # parameter validation
    if not isinstance(num_examples, int) or num_examples <= 0:
        raise ValueError(f"num_examples must be a positive integer, got {num_examples}")
    
    if not isinstance(w, torch.Tensor):
        raise TypeError(f"w must be torch.Tensor, got {type(w)}")
    
    w = w.squeeze()
    if w.dim() != 1:
        raise ValueError(f"w must be 1-D after squeeze, got {w.dim()}-D")
    d = w.numel()
    
    if device is None:
        device = w.device
    else:
        device = torch.device(device)
    
    w = w.to(device)
    if isinstance(b, torch.Tensor):
        b = b.to(device)
    else:
        b = torch.tensor(b, device=device, dtype=w.dtype)
    
    # random seed setup
    if seed is not None:
        generator = torch.Generator(device=device).manual_seed(seed)
    else:
        generator = None
    
    # generate features X ~ N(0, 1)
    X = torch.normal(
        mean=0, std=1, size=(num_examples, d),
        generator=generator, device=device, dtype=w.dtype
    )
    
    # generate labels y = Xw + b + noise
    y = torch.matmul(X, w) + b
    if noise_std > 0:
        noise = torch.normal(
            mean=0, std=noise_std, size=y.shape,
            generator=generator, device=device, dtype=w.dtype
        )
        y = y + noise
    
    y = y.reshape(-1, 1)
    return X, y


class LinearRegressionModel(nn.Module):
    """
    简单的线性回归模型，等价于 nn.Linear(input_dim, 1)
    """
    def __init__(self, input_dim: int):
        super().__init__()
        self.linear = nn.Linear(input_dim, 1)
        
        # initialize weights with small random values
        nn.init.normal_(self.linear.weight, mean=0, std=0.01)
        nn.init.zeros_(self.linear.bias)
    
    def forward(self, x: torch.Tensor) -> torch.Tensor:
        return self.linear(x)


def train_epoch(
    model: nn.Module,
    dataloader: DataLoader,
    criterion: nn.Module,
    optimizer: torch.optim.Optimizer,
    device: torch.device
) -> float:
    """
    训练一个 epoch，返回平均损失。
    """
    model.train()
    total_loss = 0.0
    num_batches = 0
    
    for batch_X, batch_y in dataloader:
        # move data to target device
        batch_X = batch_X.to(device)
        batch_y = batch_y.to(device)
        
        # forward pass
        pred = model(batch_X)
        loss = criterion(pred, batch_y)
        
        # backward pass
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
        total_loss += loss.item()
        num_batches += 1
    
    return total_loss / num_batches if num_batches > 0 else 0.0


def evaluate(
    model: nn.Module,
    dataloader: DataLoader,
    criterion: nn.Module,
    device: torch.device
) -> float:
    """
    在验证集/测试集上评估，返回平均损失。
    """
    model.eval()
    total_loss = 0.0
    num_batches = 0
    
    with torch.no_grad():
        for batch_X, batch_y in dataloader:
            batch_X = batch_X.to(device)
            batch_y = batch_y.to(device)
            
            pred = model(batch_X)
            loss = criterion(pred, batch_y)
            
            total_loss += loss.item()
            num_batches += 1
    
    return total_loss / num_batches if num_batches > 0 else 0.0


def save_checkpoint(
    model: nn.Module,
    optimizer: torch.optim.Optimizer,
    epoch: int,
    loss: float,
    filepath: str
) -> None:
    """
    保存训练检查点到 .pth 文件。
    """
    checkpoint = {
        'epoch': epoch,
        'model_state_dict': model.state_dict(),
        'optimizer_state_dict': optimizer.state_dict(),
        'loss': loss,
    }
    torch.save(checkpoint, filepath)
    print(f"Checkpoint saved to {filepath}")


def load_checkpoint(
    filepath: str,
    model: nn.Module,
    optimizer: Optional[torch.optim.Optimizer] = None,
    device: Optional[torch.device] = None
) -> dict:
    """
    从 .pth 文件加载检查点。
    """
    if device is None:
        device = torch.device('cpu')
    
    # map_location ensures loading on CPU even if saved on GPU
    checkpoint = torch.load(filepath, map_location=device)
    
    model.load_state_dict(checkpoint['model_state_dict'])
    
    if optimizer is not None and 'optimizer_state_dict' in checkpoint:
        optimizer.load_state_dict(checkpoint['optimizer_state_dict'])
    
    print(f"Checkpoint loaded from {filepath}")
    return checkpoint


def main():
    # ---------- hyperparameters ----------
    TRUE_W = torch.tensor([2.0, -3.4])
    TRUE_B = 4.2
    NUM_TRAIN = 1000
    NUM_VAL = 200
    BATCH_SIZE = 32
    LEARNING_RATE = 0.03
    NUM_EPOCHS = 10
    SAVE_DIR = './checkpoints'
    DEVICE = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    
    # create save directory
    os.makedirs(SAVE_DIR, exist_ok=True)
    
    # ---------- data generation ----------
    print("Generating synthetic data...")
    X_train, y_train = synthetic_data(TRUE_W, TRUE_B, NUM_TRAIN, seed=42)
    X_val, y_val = synthetic_data(TRUE_W, TRUE_B, NUM_VAL, seed=43)
    
    # move to device
    X_train = X_train.to(DEVICE)
    y_train = y_train.to(DEVICE)
    X_val = X_val.to(DEVICE)
    y_val = y_val.to(DEVICE)
    
    # create dataloaders
    train_dataset = TensorDataset(X_train, y_train)
    val_dataset = TensorDataset(X_val, y_val)
    
    train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
    val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False)
    
    # ---------- model setup ----------
    input_dim = TRUE_W.numel()
    model = LinearRegressionModel(input_dim).to(DEVICE)
    
    criterion = nn.MSELoss()
    optimizer = torch.optim.SGD(model.parameters(), lr=LEARNING_RATE)
    
    print(f"Training on device: {DEVICE}")
    print(f"True parameters: w = {TRUE_W.numpy()}, b = {TRUE_B:.4f}")
    
    # ---------- training loop ----------
    best_val_loss = float('inf')
    
    for epoch in range(1, NUM_EPOCHS + 1):
        train_loss = train_epoch(model, train_loader, criterion, optimizer, DEVICE)
        val_loss = evaluate(model, val_loader, criterion, DEVICE)
        
        print(f"Epoch [{epoch:02d}/{NUM_EPOCHS}] "
              f"Train Loss: {train_loss:.6f} | Val Loss: {val_loss:.6f}")
        
        # save best model
        if val_loss < best_val_loss:
            best_val_loss = val_loss
            save_path = os.path.join(SAVE_DIR, 'best_model.pth')
            save_checkpoint(model, optimizer, epoch, val_loss, save_path)
    
    # ---------- save final checkpoint ----------
    final_path = os.path.join(SAVE_DIR, 'final_model.pth')
    save_checkpoint(model, optimizer, NUM_EPOCHS, val_loss, final_path)
    
    # ---------- inspect learned parameters ----------
    learned_w = model.linear.weight.detach().cpu().numpy().flatten()
    learned_b = model.linear.bias.detach().cpu().item()
    
    print("\n" + "=" * 40)
    print("Training complete!")
    print(f"Learned w: {learned_w}")
    print(f"Learned b: {learned_b:.4f}")
    print("=" * 40)
    
    # ---------- load and verify ----------
    print("\nLoading best model for verification...")
    new_model = LinearRegressionModel(input_dim).to(DEVICE)
    new_optimizer = torch.optim.SGD(new_model.parameters(), lr=LEARNING_RATE)
    
    best_path = os.path.join(SAVE_DIR, 'best_model.pth')
    ckpt = load_checkpoint(best_path, new_model, new_optimizer, DEVICE)
    
    print(f"Loaded from epoch {ckpt['epoch']} with loss {ckpt['loss']:.6f}")
    
    # quick verification
    new_model.eval()
    with torch.no_grad():
        sample_pred = new_model(X_val[:5])
    print(f"Sample predictions (first 5): {sample_pred.cpu().numpy().flatten()}")


if __name__ == '__main__':
    main()
```
